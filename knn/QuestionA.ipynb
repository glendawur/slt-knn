```python
from knn import KNNClassifier, accuracy
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('../data/MNIST_train_small.csv', header = None)
Y_train = df.iloc[:,0]
X_train = df.iloc[:,1:]
X_train.head(5)
```

```python
df = pd.read_csv('../data/MNIST_test_small.csv', header = None)
Y_test = df.iloc[:,0]
X_test = df.iloc[:,1:]
X_test.head(5)
```

### Question a)

#### Our implementation results

```python
train = []
test = []
for k in np.arange(1, 21):
    knn = KNNClassifier(k=k, p =2.0).fit(X_train, Y_train)
    train.append(accuracy(Y_train, knn.predict(X_train)))
    test.append(accuracy(Y_test, knn.predict(X_test)))
```

```python
import matplotlib.pyplot as plt

plt.plot(np.arange(1, 21), train, label = "Train Accuracy")
plt.plot(np.arange(1, 21), test, label = "Test Accuracy")

plt.xlabel('k-neigbors')
# Set the y axis label of the current axis.
plt.ylabel('accuracy')
# Set a title of the current axes.
plt.xticks(np.arange(1, 21))
plt.grid()
# show a legend on the plot
plt.legend()
# Display a figure.

plt.show()
```

```python
for i,j,k in zip(np.arange(1,21), train, test):
    print('{} neighbors: train score {}; test score {}'.format(i, np.round(j, 3), np.round(k, 3)))
```

#### Sklearn results

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
train = []
test = []
for k in np.arange(1, 21):
    knn = KNeighborsClassifier(n_neighbors = k, p = 2).fit(X_train, Y_train)
    train.append(accuracy(Y_train, knn.predict(X_train)))
    test.append(accuracy(Y_test, knn.predict(X_test)))
```

```python
plt.plot(np.arange(1, 21), train, label = "Train Accuracy")
plt.plot(np.arange(1, 21), test, label = "Test Accuracy")

plt.xlabel('k-neigbors')
# Set the y axis label of the current axis.
plt.ylabel('accuracy')
# Set a title of the current axes.
plt.xticks(np.arange(1, 21))
plt.grid()
# show a legend on the plot
plt.legend()
# Display a figure.

plt.show()
```

```python
for i,j,k in zip(np.arange(1,21), train, test):
    print('{} neighbors: train score {}; test score {}'.format(i, np.round(j, 3), np.round(k, 3)))
```
